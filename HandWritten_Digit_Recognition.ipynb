```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as p
import tensorflow as tf
```

```python
data_t=pd.read_csv("train.csv")
```

```python
digits=data_t.iloc[:,1:]
label=data_t.iloc[:,0]
```

```python
def mod():
    model=tf.keras.Sequential([tf.keras.layers.Dense(256,input_shape=(784,),activation='relu'),tf.keras.layers.Dense(128,activation='relu'),tf.keras.layers.Dense(64,activation='relu'),tf.keras.layers.Dense(32,activation='relu'),tf.keras.layers.Dense(10,activation='softmax')])
    model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
    return model
```

```python
%%timeit -n1 -r1
with tf.device('/CPU:0'):
    
    cpu=mod()
    cpu.fit(digits,label,epochs=5) 
```

```python
%%timeit -n1 -r1
with tf.device('/GPU:0'):
    gpu=mod()
    gpu.fit(digits,label,epochs=5)
```

```python

test=pd.read_csv("test.csv")
```

##some prediction


```python
model=mod()
model.fit(digits,label,epochs=5)
final_result=(model.predict(test))
```
